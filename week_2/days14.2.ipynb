#Pivot table

```python
import pandas as pd
```

```python
df=pd.read_csv('dayss12.csv')
df
```

```python
df.pivot(index="Day", columns="City", values=["Temperature",'Windspeed'])
```

```python
df=pd.read_csv("day14.csv") #data has morning and evening temp
df
```

```python
df.pivot_table(index='City',columns="Day", values=["Temperature",'Windspeed']) #this gives average tempr of day1+day1/2
```

```python
df.pivot_table(index='City',columns="Day", values=["Temperature",'Windspeed'], margins=True) 
```
