#Detecting NA N/A and na values

```python
missing_values=["na","N/a",np.nan]  #Pandas treats np.nan as a missing value
df=pd.read_csv("check.csv", na_values=missing_values)
```

```python
df.isnull().sum() 
#Problem is solved
```

```python
df.isnull().any()  #True shows that id and age has null 
```

#Remove these values 

```python
df
```

```python
df.fillna(method='ffill') # Forward fill: most recently observed value in the same column
```

```python
df.fillna(method='bfill') # Backward fill:next observed value in the same column. 
```

```python
df.interpolate() # interpolate() here gives the value by adding upper and lower and dividing by two
```

```python
df_=df.dropna() # dropna(): used to remove missing values 
df_
```
