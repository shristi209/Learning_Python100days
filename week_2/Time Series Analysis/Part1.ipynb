# Datetimeindex & Resampling

```python
import pandas as pd
df=pd.read_csv('HistoricalQuotes.csv',parse_dates=["Date"], index_col="Date")
df.head()
```

```python
#changing integer index into datetime index
df.index
```

```python
#benefits of having datetime index
df["2019-02-24":"2019-03-24"]
```

***Retrieve Feb 2020***

```python
df["2020-02"]
```

***Find Average Price Of apple's in Stock***

```python
df['Close']=df['Close'].str.replace('$', '') #replacing $ by none
df
```

```python
df['Close']=pd.to_numeric(df['Close']) #changing Close to numeric to calculate average
```

```python
df["2020-02"].Close.mean()
```

***Plotting Yearly frequency***

```python
%matplotlib inline
df.Close.resample('Y').mean().plot()
```
