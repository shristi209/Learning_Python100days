#Concat() Dataframes

```python
import pandas as pd
students_Devi=pd.DataFrame({"Name":['Ram', 'Shyam', 'Sita', 'Krishna'],
        "Marks":['87', '78', '90', '23']})
students_Devi
```

```python
students_Nidhi=pd.DataFrame({"Name":['Rita', 'Siru', 'Simpa', 'Kriti'],
        "Marks":['90', '70', '41', '50']})
students_Nidhi
```

***Adding another data in columns***

```python
#combining two or more use concat
pd.concat([students_Devi, students_Nidhi])
```

```python
#index continues as the previous one so use ignore_index=True to continue the index
#ignore the index of the original dataframe
df=pd.concat([students_Devi, students_Nidhi], ignore_index=True)
df
```

```python
df=pd.concat([students_Devi, students_Nidhi], keys=["students_Devi","students_Nidhi"])
df
```

```python
#df.loc() : big dataframe into original dataframe
df.loc["students_Nidhi"]
```

```python
Others=pd.DataFrame({'Faculty': ['Management', 'Science', 'Science', 'Arts'],
        'Age': ['21','20','19','21']}, index=[0,2,3,1]) #we have to mention index to locate the correct data. 
Others
```

***axis=0: rows formation, axis=1: column formation, default 0***

```python
#adding columns by using axis=1
df1=pd.concat([students_Devi, Others], axis=1)
df1
```

```python
s=pd.Series(["male","male",'female','male'], name="Gender")
s
```

```python
pd.concat([df1, s], axis=1)
```
