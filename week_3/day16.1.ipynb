```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
date=[1924,1954,1984,2014]
male=[3000,5000,5500,6000]
female=[500, 1000, 1500,4500]
```

```python
position=np.arange(len(date))  #np.arange() function creates a one-dimensional array,, elements in the date list
position                       #len(date) returns the number of elements in the date list, we have 4 elements, this return the value 4
```

```python
plt.xlabel("Years")
plt.ylabel("No. of participants")
plt.title("Male And Female Olympic Participation")  #labeling 
plt.xticks(position,date)         # xticks function  is used to set or get the current tick locations and labels of the x-axis
plt.bar(position-0.1,male, width=0.2, label="Male")   # male chart is -0.1 from the date and +0.2 of the female, use to maintain position 
plt.bar(position+0.2, female, width=0.3, label="Female")

plt.legend()  #legend function automatically takes the labels provided.
```

```python
#horizontal charts using barh
plt.ylabel("Years")
plt.xlabel("No. of participants")
plt.title("Male And Female Olympic Participation") 
plt.yticks(position,date)        
plt.barh(position-0.1,male, label="Male") 
plt.barh(position+0.2, female, label="Female")

plt.legend()
```
