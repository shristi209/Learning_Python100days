#Seaborn


#Scatter Plot

```python
import seaborn as sns
import pandas as pd
```

```python
sales=pd.read_csv('sales1.csv')
sales.head()
```

```python
sns.relplot(x='Sales',y='Profit', data=sales) 
```

```python
sns.relplot(x='Sales',y='Profit', data=sales, hue="Order Priority") 
```

```python
sns.relplot(x='Sales',y='Profit', data=sales, hue="Region",style="Region") 
```

```python
sns.relplot(x='Sales',y='Profit', data=sales, hue="Discount", size="Discount", sizes=(50,150)) #size provide the range of discount from 0.00 to 0.25, sizes gives the size of circle
```

#line

```python
sns.relplot(y='Sales',x='Order Date', data=sales, kind="line")
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])     #converting order date into date time format
sales['year'] = sales['Order Date'].dt.year                   # .dt accessor with datetimelike values
sales.year.unique()                                           #extracting the unique values
```

```python
sns.relplot(x="year", y="Sales", data=sales, kind='line', ci='sd') #ci=confidence interval, it can be None, sd
```

```python
sns.relplot(x="year", y="Sales", data=sales, kind='line', estimator=None) #estimator shows exact val, csn be None, sum
```

```python
sns.relplot(x="year", y="Sales", data=sales, kind='line',hue="Ship Mode")
```

```python
sns.relplot(x="year", y="Sales", data=sales, kind='line', hue="Product Category", style="Customer Segment", ci=None)
```

```python
sns.relplot(x="year", y="Sales", data=sales, kind='line', hue="Product Category", col="Customer Segment", ci=None)
```
