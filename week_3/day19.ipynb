```python
import seaborn as sns
import pandas as pd
```

```python
sales=pd.read_csv('sales1.csv')
sales.head()
```

```python
sns.relplot(x="Sales", y="Profit", data=sales, col="Customer Segment",row="Product Category", hue="Order Priority")
```

#**Scatter**, **Swarm** **Plot**

```python
#Scatter plot, Magnitude
sns.catplot(x="Ship Mode", y= "Sales", data= sales, hue="Ship Mode")
```

```python
#jitter
sns.catplot(x="Ship Mode", y= "Sales", data= sales, hue="Ship Mode", jitter=False)
```

```python
#Swarm Plot:type of scatter plot where the data points are adjusted so that they do not overlap with each other
sns.catplot(x="Ship Mode", y= "Sales", data=sales.query("Profit<1000"), kind="swarm")
```
