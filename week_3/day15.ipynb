#Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,5,10,15,20,30]
y=[50,47,40,55,59,68]
```

```python
plt.plot(x,y)
```

```python
#line more thicker by using linewidth=6
#change color by color='green', by default=blue
plt.xlabel('Month')
plt.ylabel('Expenditure')
plt.title('Outgoing income')
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
plt.plot(x,y,'r+-.')   #+-. line 
```

***Axes,Labels, Legend, Grid***

```python
days=[1,2,3,4,5,6,7]
max_t=[70,65,48,67,30,44,55]
min_t=[20,10,15,20,18,33,30]
avg=[45,30,27,34,25,38,45]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Maximum temp')
plt.plot(days,min_t, label='Minimum temp')
plt.plot(days, avg, label="Average temp")

plt.legend(loc='best')  #locating the label
```
