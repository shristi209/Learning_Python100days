#Box Plot

```python
import seaborn as sns
import pandas as pd
```

```python
sales=pd.read_csv('sales1.csv')
sales.head(3)
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind="box")#A box plot gives: the minimum value, the first quartile, the median, the third quartile, and the maximum value.
```

The dot dot line denotes the outlier

```python
sns.catplot(x="Product Category", y="Shipping Cost", hue="Ship Mode", data=sales, kind="box")
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", row="Ship Mode", data=sales, kind="box") #row to make individual chart
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", col="Ship Mode",row="Customer Segment", data=sales, kind="box")
```
