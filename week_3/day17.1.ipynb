#Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[103,95,90,150,149,88,93,110,80,77,82,129]
plt.hist(blood_sugar,bins=4, rwidth=0.9) #by default bin is 10, changing it to 4, also we can change into parameter
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.9, color ='y')
```

```python
#2histogram
blood_sugar_men=[103,95,90,150,149,88,93,110,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.title('Analysis Of Blood Sugar')
plt.xlabel('Sugar Range')
plt.ylabel("Total no. of patients(Frequency)")
plt.hist([blood_sugar_men, blood_sugar_women],bins=4, rwidth=0.9,color=['orange','blue'], label=['Men','Women']) #orientation is used to show histogram in horizontal or vertical way
plt.legend()                                                                                                 
```
