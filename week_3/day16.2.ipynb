#Pie Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
percentage=[34, 10, 27, 3, 26]
percentage1=[25,30,5,20,20]
money_spent=["Housing","Transport","Food","Health care","Other goods and services"]
```

```python
plt.pie(percentage, labels= money_spent, autopct='%0.0f%%', explode=[0,0,0,0,0.05]) #pie func, labeling how money are spent, explode separates the pie
plt.axis("equal")
plt.title('Money Spent in Malasiya')
plt.show()  
```

```python
#ways of showing pie charts
plt.pie(percentage, labels=money_spent , autopct='%0.0f%%') 
plt.title("Money Spent in Nepal")
plt.show()

plt.pie(percentage1, labels= None, autopct='%0.0f%%' )
plt.title('Money Spent in Malasiya')
plt.legend(money_spent, loc="best", bbox_to_anchor=(1, 1)) #this creates different table for money spent
plt.show() 
```
